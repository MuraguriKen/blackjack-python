```python
# @title
class Bank:

    def __init__(self,bank_bal=0,deposit=0,bet_amt=0):

        self.deposit = deposit
        self.bank_bal = bank_bal
        self.bet_amt = bet_amt

    def __str__(self):

        print()
        return f'Available bank balance is KES {self.bank_bal}'

    def bank_initial(self):

        #self.bank_roll = 0

        while True:
            try:
                print()
                self.deposit = int(input("Enter cash to convert to deposit: "))
            except ValueError:
                print("Wrong input. Please enter a number!!")
            else:
                if self.deposit <= 0:
                    print("Wrong input. Please enter a positive integer!!")
                else:
                    break

        self.bank_bal += self.deposit

        return self.bank_bal

    def bet(self):

        self.bet_amt = 0

        while self.bet_amt == 0:
            more = " "
            try:
                print()
                self.bet_amt = int(input("Enter your bet: "))
            except:
                print("Please enter an integer")
                continue

            if self.bet_amt <= 0:
                print("Please enter an amount greater than 0")
                self.bet_amt = 0
            elif self.bet_amt <= self.bank_bal:
                break
            else:
                while self.bet_amt > self.bank_bal:
                    print(f"Bet amount is greater that your available balance: {self.bank_bal}")

                    while more not in ('Y', 'N', 'y', 'n'):
                        print()
                        more = input("Deposit more cash? (Y/N): ")

                    if more.upper() == "Y":
                        self.depo()
                        more = " "
                        continue
                    elif self.bank_bal == 0:
                        print("Thank you for playing")
                        break
                    else:
                        self.bet_amt = 0



        if (self.bet_amt > 0) and (more.upper() != "N"):
            print()
            print(f"Your bet: {self.bet_amt}")

            self.bank_bal -= self.bet_amt

            print(f"Your Bank Balance: {self.bank_bal}")

        return self.bet_amt

    def depo(self):

        self.deposit = 0

        while self.deposit == 0:
            try:
                self.deposit = int(input("Enter cash to deposit: "))
            except ValueError:
                print("Wrong input. Please enter a number!!")
            else:
                if self.deposit <= 0:
                    print("Wrong input. Please enter a positive integer!!")
                    self.deposit = 0
                    continue
                #elif self.deposit.strip() == '':
                    #print("Wrong input. Please enter a number!!")
                else:
                    break

        self.bank_bal += self.deposit

        print(f"Your Bank Balance: {self.bank_bal}")

        return self.bank_bal

    def win_bal(self):

        self.bank_bal += self.bet_amt * 2

        return self.bank_bal

    def tie_bal(self):

        self.bank_bal += self.bet_amt

        return self.bank_bal
```

```python
# @title
def instructions():

    print('''Welcome to Blackjack by Mura.

OBJECTIVE: Get a higher value than the dealer without going over 21.

TERMS:
Player - You
Dealer - Your opponent / The house
Deck - A shuffled source for random cards
Suit - Name on card (e.g. Club, Diamond)
Rank - Number on card (e.g. 3, 5, Jack, Ace)
Hand - The set of cards dealt to you (Player Hand) / the dealer (Dealer Hand)
Hit - Add a random card from the deck
Stay - Decline an extra card, hold your current hand as final for the turn
Bet - Your stake for the current hand
Bust - Hand greater than 21.

RULES & Gameplay:
1. BETS
- You must stake a value to play.
- If you lose the round, you lose your bet.
- Win = bet x 2.
- Tie = bet returned to you.

2. CARDS:
- 3 to 10 have a value equivalent to their rank.
- Jack, King and Queen have a value of 10.
- Ace: Value of 11 if total value will not exceed 21.
       Value of 1 if with a value of 11 total value will exceed 21.

3. Dealer:
- Dealt two cards.
- Only one card face up till the player completes their play.
- Dealer always hits if hand is 17 or below.
- Dealer stays if hand above 17 (18 and above).

4. Player:
- Deposits cash if no cash available.
- Places a bet.
- Dealt two cards face up.
- Must make a decision whether to hit or stay at every turn.
- If player busts, they lose, no need for dealer hand reveal.
- If player stays, dealer plays till cannot anymore.
- Hand closest to 21 wins.
- If both dealer and player hand values equal, it's a tie.

READY TO PLAY?

''')
```
